# created a new column named 'correct_shipdate' using funtion with ship_mode and order_date

```python
def right (x):
    if x['Ship_Mode'] == 'Same Day':
        return x ['Order_Date']
    elif x['Ship_Mode'] == 'Standard Class':
         return x ['Order_Date'] + pd.Timedelta (days = np.random.randint(1,4))
    elif x['Ship_Mode'] == 'First Class':
        return x ['Order_Date'] + pd.Timedelta (days = np.random.randint(3,6))
    elif x['Ship_Mode'] ==  'Second Class':
        return x ['Order_Date'] + pd.Timedelta (days = np.random.randint(6,9))
```

```python
df['Correct_shipdate'] = df.apply(right,axis=1)
```

# created new column Order_delay and shipping year for analysis purpose 

```python
df['Orders_delay']=df['Correct_shipdate'] - df['Order_Date']
```

```python
df['Shipping_years']=df['Correct_shipdate'].dt.year
```
